```python
# https://realpython.com/build-recommendation-engine-collaborative-filtering/
# https://medium.com/@17it097/movie-recommendation-system-edc72a9e7571
import numpy as np
import pandas as pd
import sklearn.metrics.pairwise as pw
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
df=pd.read_csv("../../Data/ml-latest-small/PreprocessedData_ml_latest_year_small.csv",index_col=0)
df.head()
```

```python
df= df.drop(['title', 'genres', 'year'], axis = 1)
```

```python
movies_users=df.pivot(index='movieId', columns='userId',values='rating').fillna(0)
```

```python
movies_users.head()
```

```python
#option 1 knn
# https://medium.com/@17it097/movie-recommendation-system-edc72a9e7571

mat_movies_users=csr_matrix(movies_users.values)
model_knn= NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10)
model_knn.fit(mat_movies_users)
```

```python
#option 2 ## Item Rating Based Cosine Similarity
#https://github.com/berkurka/MovieRecommender/blob/master/Movie_Recommender.ipynb

# with whom has user 3 the most similarities?

user_recomm = pw.cosine_similarity(mat_movies_users)
user_recomm_df = pd.DataFrame(user_recomm,columns=movies_users.index.values,
                 index=movies_users.index.values)
    ## Item Rating Based Cosine Similarity
#usr_cosine_df = pd.DataFrame(user_recomm_df[input_user_id].sort_values(ascending=False))
usr_cosine_df = pd.DataFrame(user_recomm_df[3].sort_values(ascending=False))
usr_cosine_df.reset_index(level=0, inplace=True)
usr_cosine_df.columns = ['userId','cosine_sim']
```
